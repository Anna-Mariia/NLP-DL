```python
def wagner_fischer(word1, word2):
    len_word1 = len(word1)
    len_word2 = len(word2)

    dist = [[0] * (len_word2 + 1) for _ in range(len_word1 + 1)]

    # Ініціалізація першого рядка та стовпця
    for i in range(len_word1 + 1):
        dist[i][0] = i
    for j in range(len_word2 + 1):
        dist[0][j] = j

    # Заповнення матриці
    for i in range(1, len_word1 + 1):
        for j in range(1, len_word2 + 1):
            if word1[i - 1] == word2[j - 1]:
                cost = 0
            else:
                cost = 1
            dist[i][j] = min(dist[i - 1][j] + 1,      
                             dist[i][j - 1] + 1,      
                             dist[i - 1][j - 1] + cost)  

    return dist[len_word1][len_word2]

word1 = "манна"
word2 = "масса"
print(f"Відстань між '{word1}' та '{word2}' становить: {wagner_fischer(word1, word2)}")
```

```python
import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from collections import Counter

# Завантаження корпусу тексту
nltk.download('gutenberg')
text = gutenberg.raw('melville-moby_dick.txt')
words = word_tokenize(text)

# Побудова мішка слів
bag_of_words = Counter(words)

print("Мішок слів (Bag-of-Words) для тексту:")
print(bag_of_words)
```
